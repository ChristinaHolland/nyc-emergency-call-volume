```python
weather_combo.to_csv('data/weather_combo.csv')
```

```python
cntrlprk_weather = weather_combo[weather_combo['NAME'].str.contains('CENTRAL PARK')].reset_index()
```

```python
cntrlprk_weather.info()
```

```python
year = [i.split('-')[0] for i in cntrlprk_weather["DATE"]]
```

```python
month = [i.split('-')[1] for i in cntrlprk_weather["DATE"]]
```

```python
day = [i.split('-')[2] for i in cntrlprk_weather["DATE"]]
```

```python
cntrlprk_weather["YEAR"] = year
```

```python
cntrlprk_weather["MONTH"] = month
```

```python
cntrlprk_weather["DAY"] = day
```

```python
cntrlprk_weather.head()
```

```python
cntrlprk_weather["TAVG_CALC"] = (cntrlprk_weather["TMAX"] + cntrlprk_weather["TMIN"]) / 2
```

```python
cntrlprk_weather.head()
```

```python
cntrlprk_weather.drop(columns=['index', 'TAVG'], inplace=True)
```
